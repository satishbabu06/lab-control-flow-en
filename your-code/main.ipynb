## Words

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story',  'island']
```

**Print every word in upper case**

```python
# your code here
for ch in words:
    print(ch.upper())
```

**Create a new list containing only words with 5 or more letters**

```python
# your code here
new_list = []
for ch in words:
    if(len(ch)>=5):
        new_list.append(ch)
print(new_list)
```

**Print the first word starting with "t"**

```python
# your code here
for ch in words:
    if ch.startswith('t'):
        print(ch)
        break
   
```

## Numbers

**Create a list containing the square of every number from 1 to 10**

```python
# your code here
square= []
for num in range(1,11):
    square.append(num ** 2)
print( square )
```

**Print a list containing the square of every odd number from 1 to 10**

```python
# your code here
squares = []
for num in range(1, 11):
    if num % 2 != 0:
        squares.append(num ** 2)
print(squares)
```

```python
squares = [num ** 2 for num in range(1, 11) if num % 2 != 0]
print(squares)
```

**Create a list with the squares of all multiples of 8 below 1000**

```python
# your code here
squares = [num ** 2 for num in range(8, 1000,8) ]
print(squares)
```

## People

```python
people = [
    {
        "name": "Juan",
        "age": 34,
        "n_kids": 2
    },
    {
        "name": "Pepe",
        "age": 27,
        "n_kids": 0
    },
    {
        "name": "Sonia",
        "age": 41,
        "n_kids": 1
    },
    {
        "name": "Lucía",
        "age": 22,
        "n_kids": 2
    },
    {
        "name": "Leo",
        "age": 55,
        "n_kids": 5
    }
]
```

**How many people are there?**

```python
# your code here
len(people)
```

**How many people have kids**?

```python
# your code here
kids = [person for person in people if person["n_kids"] > 0]
count = len(kids)
print ( count)
```

**How many kids do they have in total?**

```python
# your code here
total_kids = sum(person["n_kids"] for person in people)
print(total_kids)
```

**In a year's time, names ending with "a" will have an extra kid. Create a list of dictionaries with people's info in a year's time**

```python
# your code here
updated_people = [
    {
        "name": person["name"],
        "age": person["age"] + 1,
        "n_kids": person["n_kids"] + 1 if person["name"].endswith("a") else person["n_kids"]
    }
    for person in people
]

print(updated_people)
```
